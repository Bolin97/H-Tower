```python
import json


class INSTRUCTION:
    def __init__(self):
        self.context = None
        self.demand = {}


instruct = INSTRUCTION()
page_json = []
with open("../data/origin data/Instructions.json", "r", encoding="utf-8") as file:
    demands = json.load(file)
instructions = []
```

```python
for data in demands:
    page_source = data['page_source']
    data_json = []
    save_context = ""
    invalid = False
    for inputs in data['data']:
        context = inputs['input']
        # 不重复的
        if save_context not in context or save_context == "":
            if instruct.context is not None and invalid is False:
                instructions.append(instruct)
            instruct = INSTRUCTION()
            invalid = False
            start_index = context.find("'''") + 3
            end_index = context.find("'''", start_index)
            extracted_context = context[start_index:end_index]
            instruct.context = extracted_context
            save_context = extracted_context
        if 'key' in list(inputs['response'][0].keys()):
            length = len(inputs['response'])
            if length > 1:
                invalid = True
            for response in inputs['response']:
                instruct.demand[response['key']] = response['value']
    for dem in instructions:
        data_dict = {
            "input": dem.context,
            "response": dem.demand
        }
        data_json.append(data_dict)
    page_dict = {
        "page_source": page_source,
        "data": data_json
    }
    page_json.append(page_dict)
    instructions = []
with open("../data/origin data/instruction_demand_0.json", "w", encoding="utf-8") as f:
    json.dump(page_json, f, indent=4, ensure_ascii=False)
```

```python
import json

with open("../test/Tower data/instruction_demand.json", "r", encoding="utf-8") as file:
    demand = json.load(file)
```

```python
contexts = []
page_sources = []
for data in demand:
    page_source = data['page_source']
    page_sources.append(page_source)
    for inputs in data['data']:
        contexts.append(inputs['input'])
```

```python
with open("../test/Tower data/tree_demand_0.json", "r", encoding="utf-8") as f:
    data = json.load(f)
trees={}
for tree in data:
    page_source = tree["page_source"]
    if page_source in page_sources:
       trees[page_source] = tree["content"]
```

```python
len(trees)
```

```python
tree_json=[]
for key,value in trees.items():
    tree_dict = {
        "page_source":key,
        "content":value
    }
    tree_json.append(tree_dict)
with open("../test/tree_demand.json", "w", encoding="utf-8") as f:
    json.dump(tree_json, f, indent=4, ensure_ascii=False)
```
