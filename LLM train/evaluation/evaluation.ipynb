```python
import json

with open("ChatGLM/loss_log.json", "r", encoding="utf-8") as file:
    data = json.load(file)
```

```python

import matplotlib.pyplot as plt


epochs = [d["epoch"] for d in data]
losses = [d["loss"] for d in data]

plt.plot(epochs, losses)
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```
